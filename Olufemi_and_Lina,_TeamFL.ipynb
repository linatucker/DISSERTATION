# DATA PREPARATION

```
# we are going to import a data set from our github repository
un_data = pd.read_csv('https://raw.githubusercontent.com/ogtaylor2008837/Femi-s-Bucket/main/country_profile_variables.csv')
un_data
```

After exploring the data set, we wil now begin the data preparation process for this dataset by keeping only the columns relevant for our analysis.


```
# this codes keeps only the required columns
un_dataset = un_data[['country', 
                      'GDP: Gross domestic product (million current US$)', 
                      'GDP per capita (current US$)', 'Life expectancy at birth (females/males, years)', 
                      'Health: Total expenditure (% of GDP)']]
un_dataset
```

Now we continue the data preparation process by renaming every column with a more appropriate name for each unclear column. We will also split the 'Life expectancy at birth (females /males, years)' column into individual male and female columns.

```
# we rename the columns appropriately
un_dataset = un_dataset.rename(columns={'country' : 'Country', 
                                        'GDP per capita (current US$)' : 'GDP Per Capita (USD)', 'GDP: Gross domestic product (million current US$)' : 'GDP (million USD)', 
                                        'Health: Total expenditure (% of GDP)' : 'Total Health Expenditure (% of GDP)',})

# we split the 'Life expectancy' column into two
un_dataset[['Female Life Expectancy At Birth (YRS)', 'Male Life Expectancy At Birth (YRS)']] = un_dataset['Life expectancy at birth (females/males, years)'].str.split('/', expand = True)
un_dataset = un_dataset.drop(columns=['Life expectancy at birth (females/males, years)'])

un_dataset
```

After futher exploration of the data, it was clear we had to remove the few rows with null values as well as rows with a Total Health Expenditure less than zero (as this shows the values are inaccurate as there can be no health expenditure less than zero or in the negative). We have also changed each columns to their appropriate variable types while also replacing an invalid number "-99.0  to Nan in some columns.

```
# this removes rows with null values and values less than or equal to 0 in 'Total Health Expenditure'
un_dataset = un_dataset[un_dataset['Male Life Expectancy At Birth (YRS)'].notna()]
un_dataset = un_dataset[un_dataset['Female Life Expectancy At Birth (YRS)'].notna()]
un_dataset = un_dataset[un_dataset['Total Health Expenditure (% of GDP)'] > 0]

# Here we convert the columns types appropriately
un_dataset['Male Life Expectancy At Birth (YRS)'] = pd.to_numeric(un_dataset['Male Life Expectancy At Birth (YRS)'], errors='coerce')
un_dataset['Total Health Expenditure (% of GDP)'] = pd.to_numeric(un_dataset['Total Health Expenditure (% of GDP)'], errors='coerce')
un_dataset['Female Life Expectancy At Birth (YRS)'] = pd.to_numeric(un_dataset['Female Life Expectancy At Birth (YRS)'], errors='coerce')

un_dataset
```

After preparing the 'un_dataset', we can see the index column as been set as the 'Location Column' and we now have a final table of 187 columns and 7 rows below.


```
# sets index
un_dataset = un_dataset.set_index('Country')
un_dataset
```

We will now import another dataset for our analysis and save it in a variable named "who_covid". 

```
#This imports the dataset from our github repository
#who_covid = pd.read_csv('https://raw.githubusercontent.com/ogtaylor2008837/Femi-s-Bucket/main/owid-covid-data.csv', parse_dates=['date'])
#who_covid

who_covid = pd.read_csv('https://raw.githubusercontent.com/ogtaylor2008837/Femi-s-Bucket/main/owid-covid-data%20(2)%20updated%20march25.csv', parse_dates=['date'])
who_covid
```

After exploring the dataset, we noticed it contains values for 'World', 'Africa, 'Europe' etc in the 'location' column which is supposed to be filled with just countries. So we will drop every row whose value is not a country by using the "pycountry" library we imported earlier to create a 2 letter country ISOCODE which only generates if the name of the element in the "Location" column and row is a valid country.

```
# we define a function to create country iso code
def do_fuzzy_search(country):
    try:
        result = pycountry.countries.search_fuzzy(country)
    except Exception:
        return np.nan
    else:
        return result[0].alpha_2

iso_map = {country: do_fuzzy_search(country) for country in who_covid["location"].unique()}
who_covid["ISO code 2"] = who_covid["location"].map(iso_map)
who_covid = who_covid[who_covid['ISO code 2'].notna()]
who_covid
```

Now, we are going to pick and display only columns relevant for our analysis and also rename the columns appropriately. We will also replace the name of the country "United State" to its full name to enable us join the table easily with other tables moving forward.

```
# this selects the relevant columns
who_covid = who_covid[['date', 'iso_code', 'ISO code 2', 'continent', 'location', 'new_cases', 'new_deaths',
                       'new_tests', 'new_vaccinations','hospital_beds_per_thousand', 'population', 'aged_65_older']]

# this renames the columns
who_covid_1 = who_covid.rename(columns={'iso_code' : 'ISO_Code', 'continent' : 'Continent', 
                                      'location' : 'Country', 'date' : 'Date', 
                                      'new_cases' : 'Cases', 'new_deaths': 'Deaths', 
                                      'new_tests' : 'Tests', 'new_vaccinations' : 'Vaccinations', 
                                      'median_age' : 'Median Age', 'aged_65_older' : 'Aged 65 & Older', 
                                      'extreme_poverty' : 'Extreme Poverty', 'cardiovasc_death_race' : 'Cardiovasc Death Race', 
                                      'human_development_index' : 'Human Dev Index', 'population' : 'Population (2020)', 
                                      'population_density' : 'Population Density (2020)', 'hospital_beds_per_thousand' : 'Hospital Beds Per Thousand',
                                      'handwashing_facilities' : 'Handwashing Facilities'})

# replaces a country name
who_covid_1.replace(to_replace=['United States'], value= ['United States of America'], inplace=True)



#returns negative values to positive values using absolute values
who_covid_1
who_covid_1['Cases'] = abs(who_covid_1['Cases'])
who_covid_1

```

we are now going to import a new dataset gotten from [link text](https://www.kaggle.com/paultimothymooney/latitude-and-longitude-for-every-country-and-state). This dataset has "latitude" and "longitude" which we will need later on to plot a map chart



```
# imports the dataset
lat_lon = pd.read_csv('https://raw.githubusercontent.com/ogtaylor2008837/Femi-s-Bucket/main/world_country_and_usa_states_latitude_and_longitude_values.csv')
lat_lon = lat_lon[["country_code", "latitude", "longitude", "country"]]

lat_lon
```

We will now set the 'Country' column of the 'who_covid table as index so as to enable us join the "who_covid dataframe with the initially imported "un_data" dataframe based on their common columns which are the countries. After joining the two tables, we will then reset the index column. We will also merge the joined table with the above lon_lat table with the ISO code 2 as the common columns.

```
# sets index
who_covid_1 = who_covid_1.set_index('Country')

# this joins the 2 tables & resets the index
full_data = who_covid_1.join(un_dataset, how='inner', on='Country')
full_data = full_data.reset_index()
full_data = full_data.merge(lat_lon, how='inner', left_on='ISO code 2', right_on='country_code')
full_data










```

Now we will resample the dataframe so that the date record for each country will now be monthly instead of daily while also creating new columns for the % per population for covid cases, deaths, tests and vaccinations. This will be achieved by dividing the cases by population and then multiplying by 100.

```
# this resamples to monthly date while grouping by location
full_data = pd.DataFrame(full_data.set_index('Date').groupby('Country').resample('M').agg({'ISO_Code':'last','Continent':'last', 'Cases':'sum', 'Deaths':'sum', 
                                                                                            'Tests':'sum', 'Vaccinations':'sum', 'Aged 65 & Older':'last', 'Population (2020)':'last', 
                                                                                            'GDP (million USD)':'last', 'GDP Per Capita (USD)':'last', 
                                                                                           'Total Health Expenditure (% of GDP)':'last', 
                                                                                            'Female Life Expectancy At Birth (YRS)':'last', 'Male Life Expectancy At Birth (YRS)':'last', 
                                                                                           'latitude':'last', 
                                                                                            'longitude':'last', 'Hospital Beds Per Thousand':'last'}))
# this runs the calculations for the new columns
full_data['% Cases per Population'] = (full_data['Cases'] / full_data['Population (2020)'] * 100).round(2)
full_data['% Deaths per Population'] = (full_data['Deaths'] / full_data['Population (2020)'] * 100).round(2)
full_data['% Vaccinations per Population'] = (full_data['Vaccinations'] / full_data['Population (2020)'] * 100).round(2)
full_data['% Tests per Population'] = (full_data['Tests'] / full_data['Population (2020)'] * 100).round(2)

full_data

```

Below, we will further resample the monthly records to a 16 months record which is the entire period of the dates column in our dataframe, so as to get the total cases, deaths, tests, vaccinations as well as their % per Population values for each country.  

```
# resets the index before resampling
full_data = full_data.reset_index()
period_16months = pd.DataFrame(full_data.set_index('Date').groupby('Country').resample('16M').agg({'ISO_Code':'last','Continent':'last', 'Cases':'sum', 'Deaths':'sum', 
                                                                                            'Tests':'sum', 'Vaccinations':'sum', 'Aged 65 & Older':'last', 'Population (2020)':'last', 
                                                                                            'GDP (million USD)':'last', 'GDP Per Capita (USD)':'last', 
                                                                                           'Total Health Expenditure (% of GDP)':'last', 
                                                                                            'Female Life Expectancy At Birth (YRS)':'last', 'Male Life Expectancy At Birth (YRS)':'last', 
                                                                                           'latitude':'last', 
                                                                                            'longitude':'last', 'Hospital Beds Per Thousand':'last', '% Cases per Population':'sum', 
                                                                                            '% Deaths per Population':'sum', '% Tests per Population':'sum',
                                                                                            '% Vaccinations per Population':'sum'}))

period_16months


```

From the output above, it is clear why we chose a 16 months period which is to ensure that all values for every country are properly resampled because some countries did not start having values until March 2020 eg. Zimbabwe.
Now to display single final values for each countries we will select the last dates being the end of the 16 months period for each country.

```

period_16months = period_16months.reset_index()
period_16months= period_16months.set_index('Date')
period_16months
complete = period_16months[(period_16months.index == '2021-06-30') + (period_16months.index == '2021-07-31') + (period_16months.index == '2021-05-31')]





complete
#print(complete.to_string())

```
